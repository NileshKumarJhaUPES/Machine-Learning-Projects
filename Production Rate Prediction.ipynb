```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('42918100.csv')
```

```python
data
```

```python
data.summary()
```

```python
y = data['PRO']
x1 = data[['RT','SC','P','GS','PE','FF','FP','F','PR',]]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```
